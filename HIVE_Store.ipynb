```
def HIVE_Store(df, outputpath, tblstring, _tblName, partitionBy):
  df.repartition(partitionBy).write.partitionBy(partitionBy).option("path",outputpath + tblstring).saveAsTable(_tblName, format='parquet', mode='overwrite',option='snappy')
```
