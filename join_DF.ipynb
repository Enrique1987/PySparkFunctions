```
def join_DF(df1, df2, join_field_list, join_type):
  return df1.join(df2, join_field_list, how=join_type)
#return df1.join(df2, ['id'], how='full')
```

```
from pyspark.sql import functions as F
def join_advanced_DF(df1, df2, join_field_list1, join_field_list2, join_type):
  #first rename columns (of the second DF) so the list of fields in both data frame have the same names and easy to join
  mapping = dict(zip(join_field_list2 , join_field_list1))
  df2 = df2.select([F.col(c).alias(mapping.get(c, c)) for c in df2.columns])
  return df1.join(df2, join_field_list1, how=join_type)
```

```
from pyspark.sql import functions as F
def join_DifferentField_DF(df1, df2, join_field1, join_field2, join_type):
  return df1.withColumn("id", F.col(join_field1)).join(df2.withColumn("id", F.col(join_field2)), on="id" , how=join_type)\
    .select("*")
```
