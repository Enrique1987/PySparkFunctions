```
import pyspark.sql.functions as F
def date_comparison_col_DF(df, field1, field2, aliasNewColum, comparisonType):
  if comparisonType == "==" or comparisonType == "=":
    return df.withColumn(aliasNewColum, F.when(F.col(field1) == F.col(field2), 1).otherwise(0))
  elif comparisonType == ">" :
    return df.withColumn(aliasNewColum, F.when(F.col(field1) > F.col(field2), 1).otherwise(0))
  elif comparisonType == "<" :
    return df.withColumn(aliasNewColum, F.when(F.col(field1) < F.col(field2), 1).otherwise(0))
  elif comparisonType == "!=" :
    return df.withColumn(aliasNewColum, F.when(F.col(field1) != F.col(field2), 1).otherwise(0))
  else:
    return df
```
