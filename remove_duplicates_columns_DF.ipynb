```
def remove_duplicates_columns_DF(df):
  new_names = []
  dup_column = []
  counter = {c: -1 for c in df.columns}
  for c in df.columns:
      counter[c] += 1
      if counter[c]:
        new_c = "DUP_" + c
        dup_column.append(new_c) 
      else:
        new_c = c
      new_names.append(new_c)

  #Rename duplicate columns 
  df = df.toDF(*new_names)

  #Remove duplicate columns
  return df.select([column for column in df.columns if column not in dup_column])
```
