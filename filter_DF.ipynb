```
import pyspark.sql.functions as F
def filter_DF(df, _filter_col_name, _filter_col_value, _criteria):
  if _criteria == "==" or _criteria == "=":
    return df.filter(F.col(_filter_col_name) == _filter_col_value)
  elif _criteria == "!=":
    return df.filter(F.col(_filter_col_name) != _filter_col_value)
  elif _criteria == ">":
    return df.filter(F.col(_filter_col_name) > _filter_col_value)
  elif _criteria == "<":
    return df.filter(F.col(_filter_col_name) < _filter_col_value)
  elif _criteria == "in":
    return df.filter(F.col(_filter_col_name).isin(_filter_col_value))
  else:
    return df  
```
