```
def store_as_csv(folderpath, filename, df, *delimit):
  """Store Data frame as a single CSV file

  :folderpath: file path
  :filename: file name
  :df: dataframe
  :delimit: (OPTIONAL) delimiter used as separator for the CSV file 
  """
  
  if len(delimit) == 1 and delimit[0] == ";":
    df.repartition(1).write.format('com.databricks.spark.csv').mode('overwrite').option("header", "true").option("delimiter", ";")\
    .option("emptyValue", None).option("nullValue", None).save(folderpath+filename) #.option("timestampFormat", "MM/dd/yyyy")
  elif len(delimit) == 1 and delimit[0] == "|":
    df.repartition(1).write.format('com.databricks.spark.csv').mode('overwrite').option("header", "true").option("delimiter", "|")\
    .option("emptyValue", None).option("nullValue", None).save(folderpath+filename) #.option("timestampFormat", "MM/dd/yyyy")
  else:
    df.repartition(1).write.format('com.databricks.spark.csv').mode('overwrite').option("header", "true").option("delimiter", ",")\
    .option("emptyValue", None).option("nullValue", None).save(folderpath+filename) #.option("timestampFormat", "MM/dd/yyyy")

  x = folderpath+filename
  l = dbutils.fs.ls(x) 
  
  for row in l: 
    for file in row: 
      file.endswith(".csv")
      break

  index = file.index("part-00000")
  file2=file[index:]
  filenameNew = filename
  dbutils.fs.mv(folderpath+filename+"/"+file2, folderpath+"/"+filenameNew+".csv")
  dbutils.fs.rm(folderpath+filename, True)
```
