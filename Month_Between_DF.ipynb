```
import pyspark.sql.functions as F
```

```
def month_between_DF(df, date_Column1, date_Column2, result_colName):
  return df.withColumn(result_colName,F.round(F.months_between(F.from_unixtime(F.unix_timestamp(F.col(date_Column1), 'dd/MM/yyy')),F.from_unixtime(F.unix_timestamp(F.col(date_Column2), 'dd/MM/yyy'))),0))
```
