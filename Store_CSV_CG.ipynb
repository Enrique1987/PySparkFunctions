```
def SaveSingleCSV(df, directory, name, gzip):
  directory0 = directory + "/" + name
  if gzip == True:
    df.repartition(1).write.mode("overwrite").format("csv").option("header", "true").option("codec", "org.apache.hadoop.io.compress.GzipCodec").save(directory0)
    for file in dbutils.fs.ls(directory0):
      if file.name.find("part-00000") > -1:
        dbutils.fs.mv(file.path, directory0 + ".csv.gz")
  else:
    df.repartition(1).write.mode("overwrite").format("csv").option("header", "true").save(directory0)
    for file in dbutils.fs.ls(directory0):
      if file.name.find("part-00000") > -1:
        dbutils.fs.mv(file.path, directory0 + ".csv")
  dbutils.fs.rm(directory0, True) 
```
