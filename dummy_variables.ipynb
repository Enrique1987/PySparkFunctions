```
import pyspark.sql.functions as F 
def dummy_variables(df, fieldname):
  categ = df.select(F.col(fieldname)).distinct().rdd.flatMap(lambda x:x).collect()
  exprs = [F.when(F.col(fieldname) == cat,1).otherwise(0)\
            .alias(str(cat)) for cat in categ]
  return df.select(exprs+df.columns)
```
