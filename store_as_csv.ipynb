```
def store_as_csv(folderpath, filename, df, *delimit):
  if len(delimit) == 1 and delimit[0] == ";":
#     df.coalesce(1).write.format('com.databricks.spark.csv').mode('overwrite').option("header", "true").option("delimiter", ";").save(folderpath+filename)
    df.repartition(1).write.format('com.databricks.spark.csv').mode('overwrite').option("header", "true").option("delimiter", ";").save(folderpath+filename)
  elif len(delimit) == 1 and delimit[0] == "|":
    df.repartition(1).write.format('com.databricks.spark.csv').mode('overwrite').option("header", "true").option("delimiter", "|").save(folderpath+filename)
  else:
    df.repartition(1).write.format('com.databricks.spark.csv').mode('overwrite').option("header", "true").option("delimiter", ",").save(folderpath+filename)

  x = folderpath+filename
  l = dbutils.fs.ls(x) 
  
  for row in l: 
    for file in row: 
      file.endswith(".csv")
      break

  index = file.index("part-00000")
  file2=file[index:]
  filenameNew = filename
  dbutils.fs.mv(folderpath+filename+"/"+file2, folderpath+"/"+filenameNew+".csv")
  dbutils.fs.rm(folderpath+filename, True)
```
