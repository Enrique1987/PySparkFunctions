```
def Store_history(df, outputpath, partitionBy):
#def Store_history(df, outputpath, tblstring, _tblName, partitionBy):
  if (directory_exist(outputpath)):
    df_hist = spark.read.option("delimiter",",").load(outputpath,format="parquet", inferSchema="true", header="true")
    df_final = df.union(df_hist)
    #dbutils.fs.rm(outputpath, True)
  else:
    df_final = df
  df_final.repartition(partitionBy).write.partitionBy(partitionBy).save(outputpath, format='parquet', mode='overwrite',option='snappy') 
```
