```
from pyspark.sql.types import *
def create_schema_df(StringColumnsList, IntColumnList):
    columns_struct_fields = [StructField(field_name, StringType(), True)
                                 for field_name in StringColumnsList]
    
    for i in range(0, len(IntColumnList)):
      columns_struct_fields.append(StructField(IntColumnList[i], DoubleType(), True))
      
    return StructType(columns_struct_fields)
```
