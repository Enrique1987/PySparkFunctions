```
import pyspark.sql.functions as F

def groupBy_count(df, _field_groupby, _group_calc_field):
  #return df.groupBy(_field_groupby).count()
  return df.groupBy(_field_groupby).agg(F.count("*").alias(_group_calc_field))
```
